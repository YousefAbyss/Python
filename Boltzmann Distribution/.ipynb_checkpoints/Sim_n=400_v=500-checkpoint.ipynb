```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import animation
from matplotlib.animation import PillowWriter
plt.style.use(['science', 'notebook'])
from scipy.spatial import cKDTree
```

```python
box = [1, 1]
r = np.random.random((2,n_particles))
# Color particles the start on either side
ixr = r[0] > box[0] / 2 #right
ixl = r[0] <= box[0] / 2 #left
```

```python
v = np.zeros((2,n_particles))
v[0][ixr] = -500
v[0][ixl] = 500
```

```python
def compute_new_v(v1, v2, r1, r2):
    v1_new = v1 - np.diag((v1-v2).T@(r1-r2))/np.sum((r1-r2)**2, axis=0) * (r1-r2)
    v2_new = v2 - np.diag((v2-v1).T@(r2-r1))/np.sum((r2-r1)**2, axis=0) * (r2-r1)
    return v1_new, v2_new

def motion(r, v, ts, dt):
    rs = np.zeros((ts, r.shape[0], r.shape[1]))
    vs = np.zeros((ts, v.shape[0], v.shape[1]))
    # Initial State
    rs[0] = r.copy()
    vs[0] = v.copy()
    for i in range(1, ts):
        my_tree = cKDTree(r.T)
        pairs = np.array(sorted(list(my_tree.query_pairs(r=2*radius))))
        
        if pairs.shape != (0,):
            v[:, pairs[:,0]], v[:, pairs[:,1]] = compute_new_v(v[:, pairs[:,0]], v[:, pairs[:,1]], r[:, pairs[:,0]], r[:, pairs[:,1]])

        v[0, (r[0] + radius) > box[0]] = -np.abs(v[0, (r[0] + radius) > box[0]])
        v[0, (r[0] - radius) < 0] = np.abs(v[0, (r[0] - radius) < 0])
        v[1, (r[1] + radius) > box[0]] = -np.abs(v[1, (r[1] + radius) > box[0]])
        v[1, (r[1] - radius) < 0] = np.abs(v[1, (r[1] - radius) < 0])
        
        r = r + v*dt
        rs[i] = r.copy()
        vs[i] = v.copy()
    return rs, vs
```

```python
n_particles = 400
r = np.random.random((2,n_particles))
ixr = r[0]>0.5 
ixl = r[0]<=0.5 
v = np.zeros((2,n_particles))
v[0][ixr] = -500
v[0][ixl] = 500
radius = 0.0015
rs, vs = motion(r, v, ts=1000, dt=0.000008)
```

```python
v = np.linspace(0, 2000, 1000)
a = 2/500**2
fv = a*v*np.exp(-a*v**2 / 2)
```

```python
bins = np.linspace(0, 1500, 10)
plt.figure()
plt.hist(np.sqrt(np.sum(vs[-1]**2, axis=0)), bins=bins, density=True)
plt.plot(v,fv)
plt.xlabel('Velocity [m/s]')
plt.ylabel('# Particles')
plt.show()
```

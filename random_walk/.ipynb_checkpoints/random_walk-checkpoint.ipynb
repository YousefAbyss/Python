* Import packages

```python
import numpy as np
from scipy.stats import skew
import random
from numba import njit  # for better performance
import matplotlib.pyplot as plt
plt.style.use(['science', 'notebook', 'grid'])
```

* Task 1:

```python
@njit
def walk(n_steps):
    x = 0
    my_list = []
    for _ in range(n_steps):
        r = random.uniform(0, 1)
        if r >= 0.5:
            x += 1
        else:
            x -= 1
        my_list.append(x)
    return my_list
```

```python
steps = 20000
time = np.linspace(1, steps, steps)
walks = walk(steps)
plt.plot(time, walks)
plt.xlabel('time')
plt.ylabel('x')
plt.show()
```

* Task 2:

```python
f_list = []
n_walkers = 10000
steps = 20000
for _ in range(n_walkers):  # takes about 20 seconds to calculate
    f_list.append(walk(steps))
my_array = np.array(f_list)
```

```python
time = np.linspace(1, steps, steps)
for j in range(3):
    plt.plot(time, my_array[j])
plt.xlabel('time')
plt.ylabel('x')
plt.show()
```

* Histograms:

```python
fig, ax = plt.subplots(1, 1, figsize=(10,5))
ax.set_xlabel('$X$')
ax.set_ylabel('$Occupancy$')

ax.hist(my_array[:,:100], density=True, histtype='step', bins=50)
ax.text(0.8, 0.9, '$N = 100$', size=15, bbox=dict(facecolor='white', edgecolor='black'), transform=ax.transAxes)
plt.show()
```

### Statistics:

* Mean for N = 100, 1000, 10000, 20000:

```python
for i in [100, 1000, 10000, 20000]:
    print(f'N = {i} --> <x> =', np.mean(my_array[:,:i]))
```

* Variance for N = 100, 1000, 10000, 20000:

```python
for i in [100, 1000, 10000, 20000]:
    print(f'N = {i} --> Variance =', np.var(my_array[:,:i]))
```

```python
for i in [100, 1000, 10000, 20000]:
    print(f'N = {i} --> Skewness =', skew(np.reshape(my_array[:,:i], i * n_walkers)))
```
